```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
age = np.array([10, 12, 11, 13, 26, 28, 25, 29])
spend = np.array([2, 1, 3, 1, 6, 9, 7, 8])
```

```python
dict = {"Age":age, "Spend_k":spend}
```

```python
df1 = pd.DataFrame(dict)
```

```python
df1.head()
```

```python
plt.figure(figsize=(4,3))
plt.scatter(df1["Age"], df1["Spend_k"])
plt.xlabel("Age")
plt.ylabel("Spend")
plt.title("Age vs Spendings")
plt.show()
```

```python
df1
```

```python
# Considering k=2 as centroids
cent_1 = df1.iloc[1]
cent_1
```

```python
cent_2 = df1.iloc[4]
cent_2
```

```python
cent_1_ls = []
for i in range(8):
    pc1 = np.sqrt((cent_1[0]-df1.iloc[i][0])**2 + (cent_1[1]-df1.iloc[i][1])**2)
    cent_1_ls.append(pc1)
cent_1_ls
```

```python
cent_2_ls = []
for i in range(8):
    pc2 = np.sqrt((cent_2[0]-df1.iloc[i][0])**2 + (cent_2[1]-df1.iloc[i][1])**2)
    cent_2_ls.append(pc2)
cent_2_ls
```

```python
df1_1 = df1.copy()
```

```python
df1_1["Cent_1"] = cent_1_ls
df1_1["Cent_2"] = cent_2_ls
```

```python
df1_1
```

```python
new_c1_1 = df1_1["Age"].iloc[:4].mean()
new_c1_1
```
