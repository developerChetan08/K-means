```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

# Create A Scatter Plot b/w petal length & petal width

```python
plt.scatter(df["petal length (cm)"].iloc[:50], df["petal width (cm)"].iloc[:50])
plt.scatter(df["petal length (cm)"].iloc[50:100], df["petal width (cm)"].iloc[50:100])
plt.scatter(df["petal length (cm)"].iloc[100:], df["petal width (cm)"].iloc[100:])
plt.show()
```

# Drop sepal length & sepal width Col:-

```python
df.drop(["sepal length (cm)", "sepal width (cm)"], axis = 1, inplace=True)
```

```python
df.head()
```

# Create 3 Random Centroids Points:-

```python
first_cent_1 = df.iloc[34]
first_cent_1
```

```python
first_cent_2 = df.iloc[78]
first_cent_2
```

```python
first_cent_3 = df.iloc[119]
first_cent_3
```

# Find Distance By Euclidain Distance:-

```python
first_cent_1_ls = []
for i in range(len(df)):
    pc1 = np.sqrt((first_cent_1[0]-df.iloc[i][0])**2 + (first_cent_1[1]-df.iloc[i][1])**2)
    first_cent_1_ls.append(pc1)
first_cent_1_ls
```

```python
first_cent_2_ls = []
for i in range(len(df)):
    pc1 = np.sqrt((first_cent_2[0]-df.iloc[i][0])**2 + (first_cent_2[1]-df.iloc[i][1])**2)
    first_cent_2_ls.append(pc1)
first_cent_2_ls
```

```python
first_cent_3_ls = []
for i in range(len(df)):
    pc1 = np.sqrt((first_cent_3[0]-df.iloc[i][0])**2 + (first_cent_3[1]-df.iloc[i][1])**2)
    first_cent_3_ls.append(pc1)
first_cent_3_ls
```

# Create a New DataFrame:-

```python
df1 = df.copy()
df1
```

# Create 3 New Col and Add Centriod 1 & Centriod 2 & Centriod 2 in DataFrame:-

```python
df1["first_Cent_1"] = first_cent_1_ls
df1["first_Cent_2"] = first_cent_2_ls
df1["first_Cent_3"] = first_cent_3_ls
```

```python
df1
```

# Create First Cluster:-

```python
clu1 = []
for i in range(len(df1)):
    a = df1.iloc[i,2:]
    clu1.append(a.index[np.argmin(df1.iloc[i,2:])])
```

```python
df1["1st Cluster"] = clu1
```

```python
df1
```

# Draw 1st Centroid Point IN Scatter Plot:-

```python
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"])
plt.scatter(first_cent_1[0], first_cent_1[1], color="red")
plt.scatter(first_cent_2[0], first_cent_2[1], color="red")
plt.scatter(first_cent_3[0], first_cent_3[1], color="red")
plt.show()
```
